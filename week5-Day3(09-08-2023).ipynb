```python
from matplotlib import pyplot as plt
```

#### Bar Chart:
- with pyplot we can use bar() function to draw a bar graph.
- The bar() takes arguments that describe the layouts of the bars.
- The categories and their values are represented by first and second arguments as array.
- plt.bar()

```python
year = ['2020','2021','2022','2023']
gold_price=[35000,40000,45000,50000]
plt.bar(year,gold_price,color='g',width=0.3)
plt.xlabel('year')
plt.ylabel('gold_price')
plt.show()
```

```python
year = ['2020','2021','2022','2023']
gold_price=[35000,40000,45000,50000]
silver=[20000,30000,40000,45000]
plt.bar(year,gold_price,color='g',width=0.33,label='gold price')
plt.bar([1,2,3,4],silver,color='r',width=0.33,label='silver price')
plt.legend(loc='lower left')
plt.xlabel('year')
plt.ylabel('gold_price')
plt.show()
```

```python
x = ['Group A','Group B','Group C','Group D']
y =[10,20,30,40]
z = [20,30,25,30]
plt.bar(x,y)
plt.bar(x,z)
plt.show()
```

```python
#horizontal bar graph?
plt.barh([1,2,3,4],[10,20,30,40],color=['r','g','b','c'],edgecolor='black')
plt.show()
```

### scatter plot:
- scatter plot is a diagram where each value in data set is represented by a dot.
- plt.scatter()

```python
L1 = [1,4]
L2 = [5,9]
plt.scatter(L1,L2)
plt.show()
```

```python
import numpy as np
```

```python
x = np.random.randint(100,500,50)
y = np.random.randint(300,600,50)
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker='*')
plt.show()
```

### boxplot():
- plt.boxplot()

```python
apple=np.random.randint(1,50,10)
mango = np.random.randint(50,100,20)
banana = np.random.randint(30,50,10)
L1 =[apple,mango,banana]
plt.boxplot(L1,labels=['apple','mango','banana'])
plt.show()
```

### Histogram:
- A histogram graph showing the frequency distribution.
- It is showing the number of observations within each given interval.
- Frequency distribution of numerical data.
- plt.hist()

```python
bins = 5
plt.hist([10,5,4,5.5,11,8,8.2],bins)
plt.show()
```

```python
h = [0,0.5,0.8,2,2.51,1,2.9,3,8,7,6,5.6,10]
bins = 10
plt.hist(h,bins,rwidth=0.3)
plt.show()
```

### pie:
- plt.pie()

```python
data = ['reading','writing','playing','eating']
per =[67,50,89,77]
plt.pie(per,labels=data,colors=['c','r','b','g'],autopct='%1.3f%%',
        
       explode=[0,0,0.5,0],shadow=True)
plt.show()
```

### Area plot or stack plot:
- plt.stackplot()

```python
days = [1,2,3,4,5]
age=[45,67,50,78,34]
weights=[34,97,45,67,26]
plt.stackplot(days,age,weights,labels=['age','weights'])
plt.legend()
plt.show()
```
