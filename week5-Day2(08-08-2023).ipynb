#### Matplotlib:
- It is mainly used for the data visualization.
- matplotlib:
    - mat -> mathematical
    - plot -> charts
    - lib -> library
- axes:
    - 1. x-axis data
    - 2. y-axis data
- styles:
    markers: o-circle
             s-square
             D-dimand
             p-pentagon
             h-hexagon
- first we need to import the matplotlib library.
- syntax: 
            from matplotlib import pyplot as plt

```python
from matplotlib import pyplot as plt
```

```python
# simple line graph..
x = [10,20,30,40]
y = [4,6,8,10]
plt.plot(x,y)
# to print the graph title..
plt.title('sample line graph')
plt.show()
```

```python
# example on styles,markers,line style..
x= [10,20,30,40]
y=[1,2,3,4]
#ls = line style : dashed,solid,dotted,dashdot,none..
#lw = line width
#ms= marker size
# c=color
# mec=marker edge color
plt.plot(x,y,marker='s',ls='dashed',lw=5,ms=10,c='green',mec='yellow',mfc='red')
plt.title('marker styles')
plt.show()
```
